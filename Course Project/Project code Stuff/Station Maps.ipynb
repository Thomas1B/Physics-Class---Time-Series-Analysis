# Map of stations for minute data

```python
from myPyPackages import myData, myDates, myStats, mySignal # custom modules.
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import cm
```

```python
coastline = myData.readCoastLine()

# Station Names
station_names = ['UVicSci', 'Monterey', 'Craigflower', 'ShawniganLake','Cumberland']

# Reading in Minute data for all stations
# UVicSci_data, Monterey_data, Craigflower_data, ShawniganLake_data, Cumberland_data = [myData.readMinuteData(station) for station in station_names]
# data = [myData.readMinuteData(station) for station in station_names]
Station_info = [myData.getStationInfo(station) for station in station_names]
UVicSci_info, Monterey_info, Craigflower_info, ShawniganLake_info, Cumberland_info = Station_info
```

```python
fs = 16 # fontsize

plt.figure(figsize=(16, 12), dpi=400)
ax = plt.axes()
ax.set_facecolor('steelblue')

plt.plot(coastline.long, coastline.lati, 'k', linewidth=0.5)
plt.fill(coastline.long, coastline.lati, zorder=0, color='green')

N = len(station_names)
colors = cm.rainbow(np.linspace(0, 1, N))
for i in range(N):
    plt.plot(Station_info[i].long,
             Station_info[i].lati,
             'o', # markerstyle
             markersize = 10,
             c=colors[i],
             label=station_names[i])

plt.title("Map of Weather Stations for Minute Resolution", fontsize=fs+2, pad=10)
plt.xlabel("Longtitude [$^{\circ}E$]", fontsize=fs, labelpad=10)
plt.ylabel('Latitude [$^{\circ}N$]', fontsize=fs, labelpad=10)
plt.legend(fontsize=fs)
plt.xlim(234.9, 236.8)
plt.ylim(48.25, 49.8)

plt.tick_params(labelsize=fs-2)


# plt.savefig('../Figures/Minute/MinuteStationMap.jpeg')
# plt.close()
```

# Map of Station for Hour data

```python
station_names = ['Bowser', 'Cortes', 'Craigflower', 'Cumberland', 'HappyValley',
                 'JamesBay', 'Macaulay', 'Monterey', 'Phoenix', 'RVYC', 'Rogers', 
                 'ShawniganLake', 'Strawberry', 'UVicSci', 'VIU']

label_names = ['Bowser', 'Cortes', 'Craigflower', 'Cumberland', 'Happy Valley',
                'JamesBay', 'Macaulay', 'Monterey', 'Phoenix', 'RVYC', 'Rogers', 
                'Shawnigan Lake', 'Strawberry', 'UVic', 'VIU']
                 
Station_info = [myData.getStationInfo(station) for station in station_names]
```

```python
fs = 16 # fontsize

plt.figure(figsize=(16, 12), dpi=400)
ax = plt.axes()
ax.set_facecolor('steelblue')

plt.title("Map of Weather Stations for Hour Resolution", fontsize=fs+2, pad=10)
plt.xlabel("Longtitude [$^{\circ}E$]", fontsize=fs, labelpad=10)
plt.ylabel('Latitude [$^{\circ}N$]', fontsize=fs, labelpad=10)

plt.plot(coastline.long, coastline.lati, 'k', linewidth=0.5)
plt.fill(coastline.long, coastline.lati, color='green')

N = len(station_names)
markers = ['ok', 'k<', 'kv', 'ks', 'rD', 'mo', 'rv', 'mD', 'rs', 'm*', 'm<', 'kD', 'mv', 'k*', 'ms']
for i in range(N):
    plt.plot(Station_info[i].long,
             Station_info[i].lati, 
             markers[i],
             label = label_names[i],
             markersize = 10)

plt.legend(fontsize=fs, ncol=3)
plt.xlim(234.9, 236.75)
plt.ylim(48.25, 50.2)
plt.tick_params(labelsize=fs-2)


# plt.savefig('../Figures/Hour/HourStationMap.jpeg')
# plt.close()
```
