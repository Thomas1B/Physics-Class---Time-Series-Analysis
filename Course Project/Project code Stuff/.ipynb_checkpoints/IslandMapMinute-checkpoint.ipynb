# Map of stations

```python
from myPyPackages import myData, myDates, myStats, mySignal, myPlots # custom modules.
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from datetime import datetime as dt

# Scipy modules
import scipy.stats as sp_stats
from scipy import integrate
```

```python
# Station Names
Stations = ['UVicSci', 'Monterey', 'Craigflower', 'ShawniganLake','Cumberland']

# Filepaths
FILEPATH = '../../Data/{}'
minutepath = '../../Data/UVicWeatherdata_Minute_2022/{}_Tp.dat'

# Other Parameters. All data sets share these parameters.
START = 736330.0     # start date 
END = 738733.99931   # end date
N = 3461760          # number of data points

startStr = myDates.DateNumtoStr(START, 'min', giveTime=True)
endStr = myDates.DateNumtoStr(END, 'min', giveTime=True)

Station_info = [myData.getStationInfo(station) for station in Stations]
UVicSci_info, Monterey_info, Craigflower_info, ShawniganLake_info, Cumberland_info = Station_info
VI_coast = pd.read_csv(FILEPATH.format('VI_Coast.dat'), sep='\s+', names=['long', 'lat'])
```

```python
plt.figure(figsize=(12, 9), dpi=100)
ax = plt.axes()
ax.set_facecolor('steelblue')

plt.plot(VI_coast.long, VI_coast.lat, 'k', linewidth=0.5)
plt.fill(VI_coast.long, VI_coast.lat, zorder=0, color='green')

plt.plot(UVicSci_info.long, UVicSci_info.lati, 'ro', label='UVic')
plt.plot(Monterey_info.long, Monterey_info.lati, 'o', color='cyan', label='Monterey')
plt.plot(Craigflower_info.long, Craigflower_info.lati, 'mo', label='Craigflower')
plt.plot(ShawniganLake_info.long, ShawniganLake_info.lati, 'bo', label='Shawnigan Lake')
plt.plot(Cumberland_info.long, Cumberland_info.lati, 'o', color='orange', label='Cumberland')

plt.title("Map of Weather Stations for Minute Resolution", fontsize=14, pad=10)
plt.xlabel("Longtitude [$^{\circ}$ $W$]", fontsize=12, labelpad=10)
plt.ylabel('Latitude [$^{\circ}$ $N$]', fontsize=12, labelpad=10)
plt.legend(fontsize=12)
plt.xlim(234.9, 236.8)
plt.ylim(48.25, 49.8)


# plt.savefig('../Figures/MinuteStationMap.jpeg')
plt.show()
```
